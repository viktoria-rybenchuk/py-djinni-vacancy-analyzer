```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("PyJobScrapper/scraped_data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
english_applicants = df.groupby('English Level')['Applicants'].sum()
print("English Level and Applicants:")
english_applicants.head()
```

```python
technologies = df['Technologies']
all_technologies = []
for tech_list in technologies:
    tech_list = eval(tech_list)
    all_technologies.extend(tech_list)
technology_counts = pd.Series(all_technologies).value_counts()

plt.figure(figsize=(10, 6))

technology_counts.plot(kind='bar')

plt.title('Common Technologies for Python')
plt.xlabel('Technology')
plt.ylabel('Count')

plt.show()
```

```python
english_level_counts = df.groupby('English Level')['Job Title'].count()

plt.figure(figsize=(8, 8))

english_level_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Percentage of Vacancies by English Level')

plt.axis('equal')

plt.show()
```

```python
english_level_counts = df.groupby('Experience')['Job Title'].count()


plt.figure(figsize=(8, 8))

english_level_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Percentage of Vacancies by Experience')

plt.axis('equal')

plt.show()
```

```python
selected_columns = ['Experience', 'Viewers', 'Applicants']
selected_data = df[selected_columns]
selected_data.corr()
```

```python
print("In summary, the correlation analysis reveals that higher levels of experience tend to be associated with lower numbers of viewers and applicants. Conversely, higher numbers of viewers and applicants are correlated with more experienced candidates. These findings highlight the interplay between experience, viewership, and applicant numbers in the given context.")
```

```python
company_counts = df['Company Name'].value_counts().head(10)

plt.figure(figsize=(8, 8))

company_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Percentage of Vacancies by Company (Top 10)')

plt.axis('equal')

plt.show()
```
